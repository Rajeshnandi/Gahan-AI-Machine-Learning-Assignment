## Now Here We Perform K-means Clustering Algorithm for Knowing Target Variable and Understanding Much Better Way

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',
                    random_state = 0)
    kmeans.fit(main_dataset)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 2, init = 'k-means++', 
                random_state = 0)
y_kmeans = kmeans.fit_predict(main_dataset)
```

## Data Preprocessing
Split DataFrame in train and test

```python
x = main_dataset.drop("target",axis=1)
x.head(3)
```

```python
y = main_dataset["target"]
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## Feature Scaling
Converting different units and magnitude data in one unit.

```python
from sklearn.preprocessing import StandardScaler
scaling = StandardScaler()
x_train_scaling_data = scaling.fit_transform(x_train)
x_test_scaling_data = scaling.transform(x_test)
```

## Machine Learning Model Building
We have clean data to build the Ml model. But which Machine learning algorithm is best for the data we have to find. The output is a categorical format so we will use supervised classification machine learning algorithms.

To build the best model, we have to train and test the dataset with multiple Machine Learning algorithms then we can find the best ML model. So letâ€™s try.

First, we need to import the required packages.

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

## Logistic Regression

## Trained  Model with out Standard Scaled Data

```python
# Trained with out Standard Scaled Data

from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(x_train,y_train)
y_pred_logistic_model = logistic_model.predict(x_test)
accuracy_score(y_pred_logistic_model,y_test)
```

```python
logistic_model.score(x_train,y_train)
```

```python
logistic_model.score(x_test,y_test)
```

## So Our Logistic Regression With Out Trained Standard Scale Data  Model gives us 100 % Accuracy in our Training Data and gives 98.90% Accuracy in our Testing Data

## Now Trained Model with Standard Scaled Data

```python
# Trained with Standard Scaled Data

logistic_model2 = LogisticRegression()
logistic_model2.fit(x_train_scaling_data,y_train)
y_pred_logistic_model2 = logistic_model2.predict(x_test_scaling_data)
accuracy_score(y_test,y_pred_logistic_model2)
```

```python
logistic_model2.score(x_train_scaling_data,y_train)
```

```python
logistic_model2.score(x_test_scaling_data,y_test)
```

## So Our Logistic Regression With Trained Standard Scale Data Model gives us 100 % Accuracy in our Training Data and gives 98.29% Accuracy in our Testing Data

```python
y_pred = logistic_model.predict(x_test)
Accuracy= confusion_matrix(y_test,y_pred)
Accuracy
```

```python
cm = confusion_matrix(y_test,y_pred)
plt.title('Heatmap of Confusion Matrix', fontsize = 15)
sns.heatmap(cm, annot = True)
plt.show()
```
